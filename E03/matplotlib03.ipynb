```python
import matplotlib.pyplot as plt
```

```python
with open("/home/exam1/Desktop/data.txt") as f:
    data = f.read().strip()
    
data = data.split('\n')

x = [row.split()[0] for row in data]
y = [row.split()[1] for row in data]
plt.plot(x, y)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```
